```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import draw

def generate_pot_form_50(height=300, width=300):
    # Define the basic shape of form 50
    rim_radius = np.random.randint(80, 100)

    # Create the image (initialize to zeros - black background)
    img = np.zeros((height, width), dtype=np.uint8)

    # Draw a single ellipse (just the rim part)
    rr, cc = draw.ellipse(height//4, width//2, rim_radius, rim_radius//3)
    rr, cc = np.clip(rr, 0, height-1), np.clip(cc, 0, width-1)  # Ensure coordinates stay within bounds
    img[rr, cc] = 255  # Set the pixel values to white (255)

    return img

# Generate and display an example
pot_image = generate_pot_form_50()
plt.imshow(pot_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import zipfile
import os
import cv2
import numpy as np
from skimage.metrics import structural_similarity as ssim

# Paths to ZIP files
zip1_path = "/content/Pottery_images.zip"  # Real dataset
zip2_path = "/content/pot_images.zip"  # Generated dataset

# Function to extract images from ZIP
def extract_images(zip_path, resize_dim=(256, 256)):
    images = []
    try:
        with zipfile.ZipFile(zip_path, 'r') as zip_file:
            file_list = [f for f in zip_file.namelist() if f.endswith(('.png', '.jpg', '.jpeg'))]
            for file_name in file_list:
                try:
                    with zip_file.open(file_name) as img_file:
                        img = cv2.imdecode(np.frombuffer(img_file.read(), np.uint8), cv2.IMREAD_COLOR)
                        if img is not None:
                            img = cv2.resize(img, resize_dim)
                            images.append(img)
                        else:
                            print(f"Warning: Failed to load image {file_name}")
                except Exception as e:
                    print(f"Error processing image {file_name}: {e}")
    except Exception as e:
        print(f"Error extracting from {zip_path}: {e}")
    return images

# Load images from both ZIPs
images1 = extract_images(zip1_path)
images2 = extract_images(zip2_path)

print(f"Loaded {len(images1)} images from {zip1_path}")
print(f"Loaded {len(images2)} images from {zip2_path}")

# Compare image datasets
def compare_images(img_set1, img_set2):
    ssim_scores = []
    for i, (img1, img2) in enumerate(zip(img_set1, img_set2)):
        try:
            # Convert images to grayscale for SSIM
            gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
            gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

            # Check for zero-variance images
            if np.var(gray_img1) == 0 or np.var(gray_img2) == 0:
                print(f"Skipping comparison for pair {i} due to zero variance.")
                continue

            # Compute SSIM
            score = ssim(gray_img1, gray_img2)
            ssim_scores.append(score)
        except Exception as e:
            print(f"Error comparing images at index {i}: {e}")
    return np.mean(ssim_scores) if ssim_scores else None

# Ensure both datasets have the same number of images for comparison
num_images = min(len(images1), len(images2))
images1, images2 = images1[:num_images], images2[:num_images]

# Compute similarity
average_ssim = compare_images(images1, images2)
if average_ssim is not None:
    print(f"Average SSIM between datasets: {average_ssim}")
else:
    print("No valid image pairs for comparison.")
```

```python
from skimage.metrics import structural_similarity as ssim
from sklearn.metrics import mean_squared_error
import numpy as np
import cv2

# Function to evaluate the similarity between real and generated image datasets
def evaluate_images(img_set1, img_set2):
    ssim_scores = []
    mse_scores = []
    psnr_scores = []

    for i, (img1, img2) in enumerate(zip(img_set1, img_set2)):
        try:
            # Convert images to grayscale for SSIM and MSE comparison
            gray_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
            gray_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

            # SSIM: Compute Structural Similarity Index
            ssim_score = ssim(gray_img1, gray_img2)
            ssim_scores.append(ssim_score)

            # MSE: Compute Mean Squared Error
            mse_score = mean_squared_error(gray_img1.flatten(), gray_img2.flatten())
            mse_scores.append(mse_score)

            # PSNR: Compute Peak Signal-to-Noise Ratio
            # Convert to float to avoid negative values
            img1_float = np.float32(img1)
            img2_float = np.float32(img2)
            psnr_score = cv2.PSNR(img1_float, img2_float)
            psnr_scores.append(psnr_score)

        except Exception as e:
            print(f"Error comparing images at index {i}: {e}")

    # Compute the average scores for SSIM, MSE, and PSNR
    avg_ssim = np.mean(ssim_scores) if ssim_scores else None
    avg_mse = np.mean(mse_scores) if mse_scores else None
    avg_psnr = np.mean(psnr_scores) if psnr_scores else None

    # Print evaluation metrics
    print(f"Average SSIM: {avg_ssim:.4f}")
    print(f"Average MSE: {avg_mse:.4f}")
    print(f"Average PSNR: {avg_psnr:.4f}")

# Example of how to call the function after loading the datasets
# Ensure both datasets have the same number of images for comparison
num_images = min(len(images1), len(images2))
images1, images2 = images1[:num_images], images2[:num_images]

# Evaluate the datasets
evaluate_images(images1, images2)
```
